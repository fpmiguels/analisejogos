# Cálculo da Matriz de Correlação

```python
correlacao = sales.select_dtypes(include=['float64', 'int64']).corr()

correlacao
```

```python
plt.figure(figsize=(14, 6))
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
```
